### Lecture de mes données

```python
import pandas as pd 

popCommune = pd.read_excel('Data/population_communes.xlsx', engine="openpyxl")
refGeo = pd.read_excel('Data/referentiel-geographique.xlsx')
valeursFon = pd.read_excel('Data/valeurs-foncieres.xlsx')
valeursFon2 = pd.read_excel('Data/valeurs-foncieres-2.xlsx')

popCommune.info()
```

```python
# Lire la colonne et réencoder
def corriger_encodage(texte):
    if isinstance(texte, str):  # Vérifie que la valeur est une chaîne
        return texte.encode('cp1252').decode('utf-8')
    return texte

# Appliquer sur une colonne spécifique ou sur tout le DataFrame
popCommune['COMMUNE'] = popCommune['COMMUNE'].apply(corriger_encodage).str.upper()
popCommune['REGION'] = popCommune['REGION'].apply(corriger_encodage).str.upper()
```

```python
def corriger_encodage(texte):
    if isinstance(texte, str):  # Vérifie que la valeur est une chaîne
        return texte.encode('cp1252').decode('utf-8')
    return texte

# Appliquer sur une colonne spécifique ou sur tout le DataFrame
refGeo['reg_nom'] = refGeo['reg_nom'].apply(corriger_encodage).str.upper()
refGeo['dep_nom'] = refGeo['dep_nom'].apply(corriger_encodage).str.upper()
```

```python
#Jointure données popCommune et refGeo
df_merged = pd.merge(popCommune, refGeo, left_on='COMMUNE', right_on='com_nom', how='inner')
df_merged
```

### Création des tables de dimension et de fait

```python
#Création table de dimension Region
dim_region = df_merged[['REGION']].drop_duplicates().reset_index(drop=True)
dim_region['Id_region'] = dim_region.index + 1
col = ['Id_region'] + [col for col in dim_region.columns if col != 'Id_region']
dim_region = dim_region[col]
df_merged = df_merged.merge(dim_region, on=['REGION'], how='left')
```

```python
#Création table de dimension Population
dim_population = df_merged[['dep_code', 'com_code', 'PTOT']].drop_duplicates().reset_index(drop=True)
dim_population['Id_commune_pop'] = dim_population.index + 1
col = ['Id_commune_pop'] + [col for col in dim_population.columns if col != 'Id_commune_pop']
dim_population = dim_population[col]
df_merged = df_merged.merge(dim_population, on=['dep_code', 'com_code', 'PTOT'], how='left')
```

```python
#Création table de dimension Commune
dim_commune = df_merged[['COMMUNE', 'com_code', 'dep_code', 'dep_nom', 'Id_region', 'Id_commune_pop']].drop_duplicates().reset_index(drop=True)
dim_commune['Id_commune'] = dim_commune.index + 1
col = ['Id_commune'] + [col for col in dim_commune.columns if col != 'Id_commune']
dim_commune = dim_commune[col]
df_merged = df_merged.merge(dim_commune, on=['COMMUNE', 'com_code', 'dep_code', 'dep_nom', 'Id_region', 'Id_commune_pop'], how='left')
```

### Création d'une BDD et import des données

```python
from sqlalchemy import create_engine, Column, Integer, String, Float
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker

# Définir la base
Base = declarative_base()
```

```python
class Region(Base):
    __tablename__ = 'region'

    id = Column(Integer, primary_key=True, autoincrement=True)
    nom_region = Column(String(50), nullable=False)

class Commune(Base):
    __tablename__ = 'commune'

    id = Column(Integer, primary_key=True, autoincrement=True)
    nom = Column(String(50), nullable=False)

# Connexion à la base MySQL
DATABASE_URL = "mysql+mysqlconnector://user:password@localhost:3306/ma_base"
engine = create_engine(DATABASE_URL)

# Création d'une session
Session = sessionmaker(bind=engine)
session = Session()
```
